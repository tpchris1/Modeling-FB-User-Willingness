## Split into 3 majors
1. “文”，“語”，“經”，“教”，“管”，“藝”，“人”，“社”，“餐”，“會”，“統” ->`social`
2. “資”，“工”，“機”，“科”，“電”，“化” -> `science`
3. 其他 -> `others`

```python
get_col_names = pd.read_csv("output_all_utf8.csv",encoding="utf-8",nrows=0)
col_names = (get_col_names.columns)
myd = dict.fromkeys(col_names, eval)
output_all_old = pd.read_csv("output_all_utf8.csv",encoding="utf-8",converters=myd)
output_all_old
```

```python
social_names = ["文","語","經","教","管","藝","人","社","餐","會","統"]
science_names = ["資","工","機","科","電","化"]

social = []
science = []
others = []
majors = []
for i in preprocessed["edu_detail"]:
    if any(ele in i for ele in social_names) == True:
        majors.append("Major-Social")
    elif any(ele in i for ele in science_names) == True:
        majors.append("Major-Science")
    else:
        majors.append("Major-Others")

majors_categorized = pd.get_dummies(majors)
preprocessed = pd.concat([preprocessed, majors_categorized], axis=1, sort=False)
preprocessed
```

### 把LABEL放到最後面

```python
col_names = list(preprocessed)
last = len(col_names)
col_names.insert(len(col_names)-1,col_names.pop(col_names.index('LABEL')))
preprocessed = preprocessed[col_names]
preprocessed
```
